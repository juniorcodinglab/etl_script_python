```python
dados_csv = []

with open(path_csv, 'r') as file:
    spamreader = csv.DictReader(file, delimiter=',')
    ## spamreader = É Do tipo _csv.reader, esse é um tipo especial que vai guardar a leitura dos dados.
    for row in spamreader:
         dados_csv.append(row)


print(dados_csv)
```

```python
dados_csv[1]
```

```python
key_mapping = {
    'Nome do Item': 'Nome do Produto',
    'Classificação do Produto': 'Roupas',
    'Valor em Reais (R$)': 'Preço do Produto (R$)',
    'Quantidade em Estoque': 'Quantidade em Estoque',
    'Nome da Loja': 'Filial',
    'Data da Venda': 'Data da Venda',              
}

key_mapping
```

```python
new_dados_csv = []

for old_dict in dados_csv:
    dict_temp = {}
    for old_key, value in old_dict.items():
        dict_temp[key_mapping[old_key]] = value
    new_dados_csv.append(dict_temp)
    
new_dados_csv[0]
```

```python
combined_list = []
combined_list.extend(new_dados_csv)
combined_list.extend(dados_json)

# Juntando os 2 dados

len(combined_list)

combined_list

```

```python
nomes_colunas = list(combined_list[0].keys())

nomes_colunas
```

```python
path_dados_combinados = "../data_processed/dados_combinados.csv"

with open(path_dados_combinados, 'w') as file:
    writer = csv.DictWriter(file, fieldnames=nomes_colunas)
    writer.writeheader()
    
    for row in combined_list:
        writer.writerow(row)
```

Utilizando o método writer.writerows, o código está preparado para escrever uma lista de dicionários no arquivo CSV. Além disso, ele inclui um cabeçalho para contextualizar os dados.

```python
dados_combinados_table = [nomes_colunas]

for row in combined_list:
    linha = []
    for coluna in nomes_colunas:
        linha.append(row.get(coluna, "Indisponível"))
        
    dados_combinados_table.append(linha)

dados_combinados_table
```

```python
path_dados_combinados = '../data_processed/dados_combinados.csv'

with open(path_dados_combinados, 'w') as file:
    writer = csv.writer(file)
    # Intera por todas as linnhas
    writer.writerows(dados_combinados_table)
```
